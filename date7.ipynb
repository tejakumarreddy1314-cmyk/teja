## Data science modeling

```r
#input as the numeric
age <-as.numeric(readline(prompt = "enter your age :"))
print(age)
```

```r
#input as the text
name <-readline(prompt = "enter your name :")
print(name)
```

```r
age <-as.integer(readline(prompt = "enter your age :"))
print(age)
```

```r
#multiple variables using the cat function
name <-readline(prompt = "enter your name :")
age <-as.numeric(readline(prompt = "enter your age :"))
cat("name:", name,"\n")
cat("age:",age)
```

### If-Else If-Else Statement Example (R's equivalent of `elif`)

```r
score <- 85

if (score >= 90) {
  print("Grade A")
} else if (score >= 80) {
  print("Grade B")
} else if (score >= 70) {
  print("Grade C")
} else {
  print("Grade D or F")
}
```

```r
marks <-as.numeric(readline("enter your marks:"))
if(marks >= 90){
  print("pass")
} else if(marks<=30){
  print("fail")
}
```

```r
my_list <-list(
  name = "santhosh",
  age = 20,
  marks = c(85,90,89),
  passed = TRUE
)
my_list
```
